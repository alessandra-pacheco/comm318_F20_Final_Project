# Data Analysis for Customs and Border Protection Agents


- Data File Name: sector_profile_1960_2018.csv

CSV file has:


- border patrol staff numbers (in number of agents) by sector for southern border by fiscal year. Also has aggregated row for southern border. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
swborder_checkpoints_df = pd.read_csv('../data/sector_profile_1960_2018.csv')
```

```python
swborder_agents_df = swborder_checkpoints_df
swborder_agents_df
swborder_agents_df['number_of_agents'].max()

```

```python
#plotting line graph showing changes in agents by sector over fiscal years 
swborder_agents_df.index = swborder_agents_df['fiscal_year'].astype(int)
tsplot=swborder_agents_df.groupby('sector')['number_of_agents'].plot(figsize=(12,4), 
                                                     legend=True, ylabel='Number of agents', xlabel='Fiscal Year', title='Figure 2: Increases in Customs and Border Patrol Agents by Sector over Fiscal Years, 1992-2018')
```

## Observations: 

- Big Bend and Del Rio had a huge spike in border agents in 2005. 

 - This is also paralleled with the southwest border aggregated line, where it also spikes in 2005. 
 - Southwest border declines in staffers in 2014/2015

- Tucson is the sector with the highest allocation of border agents
